```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.distributions.empirical_distribution import ECDF

plt.style.use("../config/custom_plots.mplstyle")
```

```python
def load_degrees(l, m, p):
    path = f'../simulations/degree_test/{l}_deg_m={m}_p={p}_size=100000.txt'
    return pd.read_csv(path, header=None).values.reshape(-1,)

def load_degrees2(l, m, p):
    path = f'../simulations/degree_test/{l}_deg_m={m}_p={p}_size=100000_more_links.txt'
    return pd.read_csv(path, header=None).values.reshape(-1,)


l1_m3_p0, l2_m3_p0 = load_degrees('l1', 3, '0'), load_degrees('l2', 3, '0')
l1_m3_p1, l2_m3_p1 = load_degrees('l1', 3, '1'), load_degrees('l2', 3, '1')
l1_m10_p0, l2_m10_p0 = load_degrees('l1', 10, '0'), load_degrees('l2', 10, '0')
l1_m10_p1, l2_m10_p1 = load_degrees('l1', 10, '1'), load_degrees('l2', 10, '1')
```

```python
def plot_degree(deg, label, marker, color, NBINS=20):
    probs, bins = np.histogram(np.log10(deg), density=True, bins=NBINS)
    plt.scatter(bins[:-1], probs / probs.sum(), marker=marker, color=color, label=label)


def get_ccdf(data):
    ecdf = ECDF(data)
    return ecdf.x, 1 - ecdf.y
```

```python
LABELS = ['L1 m=3, p=0', 'L2 m=3, p=0', 'L1 m=3, p=1', 'L2 m=3, p=1']
COLORS = ['red', 'green', 'blue', 'orange']
MARKERS = ['o', 's', 'v', 'd']


VALS = [l1_m3_p0, l2_m3_p0, l1_m3_p1, l2_m3_p1]

plt.figure(figsize=(8, 5))
plt.subplots_adjust(right=1.25)
plt.subplot(1, 2, 1)
for i, deg in enumerate(VALS):
    plot_degree(deg, label=LABELS[i], color=COLORS[i], marker=MARKERS[i])
    
plt.yscale('log')
plt.ylim(1e-5, 1.1)
plt.xlabel('log(k)')
plt.ylabel('P(k)')

plt.subplot(1, 2, 2)
for i, deg in enumerate(VALS):
    x, y = get_ccdf(deg)
    plt.plot(x, y, label=LABELS[i], color=COLORS[i])

plt.legend(loc='upper right')    
plt.yscale('log')
plt.xscale('log')
plt.xlabel(r'$k$')
plt.ylabel(r'$P_c(k)$')
plt.suptitle(r"$N=10^{5}$ with 3000 additional links", fontsize=25)
```

```python
LABELS = ['L1 m=10, p=0', 'L2 m=10, p=0', 'L1 m=10, p=1', 'L2 m=10, p=1']
VALS = [l1_m10_p0, l2_m10_p0, l1_m10_p1, l2_m10_p1]

plt.figure(figsize=(8, 5))
plt.subplots_adjust(right=1.25)
plt.subplot(1, 2, 1)
for i, deg in enumerate(VALS):
    plot_degree(deg, label=LABELS[i], color=COLORS[i], marker=MARKERS[i])
    
plt.yscale('log')
plt.ylim(1e-5, 1.1)
plt.xlabel('log(k)')
plt.ylabel('P(k)')

plt.subplot(1, 2, 2)
for i, deg in enumerate(VALS):
    x, y = get_ccdf(deg)
    plt.plot(x, y, label=LABELS[i], color=COLORS[i])

plt.legend(loc='upper right')
plt.yscale('log')
plt.xscale('log')
plt.xlabel(r'$k$')
plt.ylabel(r'$P_c(k)$')
plt.suptitle(r"$N=10^{5}$ with 10000 additional links", fontsize=25)
```
